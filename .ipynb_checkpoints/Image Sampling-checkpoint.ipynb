## Machine Learning Waste Detection: Image Sampling

In this project, we will fine-tune an object detection model for identifying pieces of trash in images. 

We then use the fine-tuned model to identify pieces of trash in images of the streets and drainage canals in Quelimane, Mozambique, in order to understand whether a community-level intervention has helped to reduce the risk of flooding in the city.



We have a sample of around 8,500 unlabelled images available to help with fine-tuning. In this notebook, we take a random sample of 100 of those images which we will annotate by hand. They will be combined with another existing dataset and used for fine-tuning.

#### Import relevant libraries

```python
from PIL import Image
import random 
from matplotlib import image
from matplotlib import pyplot as plt
import os
```

#### Load my images from the folder

```python
loaded_images = []
all_files = os.listdir("/Users/gautamchauhan/Desktop/MAPS Summer Internship/Images")
# Filter out hidden files in the directory which are not images
image_files = [f for f in all_files if not f.startswith('.') and f.lower().endswith('.jpg')]

for filename in image_files:
        img = image.imread('/Users/gautamchauhan/Desktop/MAPS Summer Internship/Images/' + filename)
        loaded_images.append((img,filename))
```

```python
# Set seed for reproducibility
random.seed(4)
images_for_annotation = random.sample(loaded_images, 100)
```

```python
for pair in images_for_annotation:
    img = pair[0]
    filename = pair[1]
    plt.imsave('/Users/gautamchauhan/Desktop/MAPS Summer Internship/Annotated Images/' + filename, img)
```
